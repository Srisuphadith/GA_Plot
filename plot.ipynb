```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("CSV_Algorithm_241211_144458.csv")
```

```python
df.head(5)
```

```python
def plot_algo(df:pd,algo :str):
    y_time = []
    algorithm = df[df['Algorithms'] == algo]
    for i in algorithm['TimeTaken']:
        if i != -1:
            y_time.append(float(i))
        else:
            break
    x_param = (algorithm['Parameters'])[0:len(y_time)]
    y_time = np.array(y_time)
    x_param = np.array(x_param)
    plt.plot(x_param, y_time)

    plt.title(f"{algo} Algorithm (5-{x_param[-1]} Parameters)")
    plt.xlabel(f"Data Quantity")
    plt.ylabel("TimeTaken(second)")
    plt.savefig(f"{algo}_graph.png")
    plt.show() 
    return x_param,y_time
```

```python
x_BruteForce = []
y_BruteForce = []

x_DecreaseAndConquer = []
y_DecreaseAndConquer = []

x_DivideAndConquer = []
y_DivideAndConquer = []

x_TransformAndConquer = []
y_TransformAndConquer = []

x_SpaceAndTimeTradeOffs = []
y_SpaceAndTimeTradeOffs = []

x_DynamicProgramming = []
y_DynamicProgramming = []

x_IterativeImprovement = []
y_IterativeImprovement = []

x_Greedy = []
y_Greedy = []

x_GeneticAlgorithm = []
y_GeneticAlgorithm = []
```

```python
x_BruteForce,y_BruteForce = plot_algo(df,'BruteForce')
x_DecreaseAndConquer,y_DecreaseAndConquer = plot_algo(df,'DecreaseAndConquer')
x_DivideAndConquer,y_DivideAndConquer = plot_algo(df,'DivideAndConquer')
x_TransformAndConquer,y_TransformAndConquer = plot_algo(df,'TransformAndConquer')
x_SpaceAndTimeTradeOffs,y_SpaceAndTimeTradeOffs = plot_algo(df,'SpaceAndTimeTradeOffs')
x_DynamicProgramming,y_DynamicProgramming = plot_algo(df,'DynamicProgramming')
x_IterativeImprovement,y_IterativeImprovement = plot_algo(df,'IterativeImprovement')
x_Greedy,y_Greedy = plot_algo(df,'Greedy')
x_GeneticAlgorithm,y_GeneticAlgorithm = plot_algo(df,'GeneticAlgorithm')
```

```python
plt.plot(x_BruteForce, y_BruteForce,label = 'BruteForce')
plt.plot(x_DecreaseAndConquer, y_DecreaseAndConquer, label = 'DecreaseAndConquer')
plt.plot(x_DivideAndConquer, y_DivideAndConquer, label = 'DivideAndConquer')
plt.plot(x_TransformAndConquer, y_TransformAndConquer,label = 'TransformAndConquer')
plt.plot(x_SpaceAndTimeTradeOffs, y_SpaceAndTimeTradeOffs,label = 'SpaceAndTimeTradeOffs')
plt.plot(x_DynamicProgramming, y_DynamicProgramming, label = 'DynamicProgramming')
plt.plot(x_IterativeImprovement, y_IterativeImprovement,label = 'IterativeImprovement')
plt.plot(x_Greedy, y_Greedy,label = 'Greedy')
plt.plot(x_GeneticAlgorithm, y_GeneticAlgorithm,label = 'GeneticAlgorithm')
plt.legend()
plt.title("Compare various Algorithm (5-100 Parameters)")
plt.xlabel('Data Quantity')
plt.ylabel('TimeTaken(second)')
#plt.grid()
plt.savefig("Compare_various_Algorithm.png")
plt.show()
```

```python
ga_val = (df[df['Algorithms'] == 'GeneticAlgorithm'])['MaxCost']
ga_val = [int(x) for x in ga_val]
ga_val = np.array(ga_val)

dp_val = (df[df['Algorithms'] == 'DynamicProgramming'])['MaxCost']
dp_val = [int(x) for x in dp_val]
dp_val = np.array(dp_val)
x = range(5,101)
x = np.array(x)
plt.plot(x,ga_val,'x',label = 'GA_MaxCost')
plt.plot(x,dp_val,label = 'DP_MaxCost')
plt.legend()
plt.xlabel('Data Quantity')

plt.ylabel('MaxCost')
plt.title('compare maxcost GA and DP')
#plt.grid()
plt.savefig("compare maxcost GA and DP.png")
plt.show()
```
